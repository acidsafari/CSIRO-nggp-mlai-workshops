```python
import subprocess
import sys
from pathlib import Path

REPO_URL = "https://github.com/nextgenerationgraduatesprogram/nextgen2026-mlai-workshops.git"
BRANCH = "workshop0-setup"
PROJECT_DIR = Path("/content/nextgen2026-mlai-workshops")

if "google.colab" in sys.modules:
    if not PROJECT_DIR.exists():
        subprocess.run(
            ["git", "clone", "--depth", "1", "--branch", BRANCH, REPO_URL, str(PROJECT_DIR)],
            check=True,
        )
    subprocess.run([sys.executable, "-m", "pip", "install", "-q", "-e", str(PROJECT_DIR)], check=True)
```
